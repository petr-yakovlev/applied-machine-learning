```python
# import stopwords and punctuation
from nltk.corpus import stopwords # Importing list of stop words from nltk
from string import punctuation # Importing list of punctuations from string
punctuations=list(punctuation)
stopWords=stopwords.words('english')
stopWords+=punctuations # Combined list of stop words
```

```python
# set the Hashing vectorizer (number of fetures is high in order to eliminate possible hash collision)
vectorizer = HashingVectorizer(n_features=10000, norm=None, stop_words=stopWords)
```

```python
# perform vectorizer
vectors = vectorizer.fit_transform(data.data)
```

```python
print(vectors[0])
```

```python
# train/test split and KFold

num_folds = 10
kfold = KFold(n_splits=num_folds, shuffle=True, random_state=102)

X_train, X_test, y_train, y_test = train_test_split(vectors, data.target, test_size=0.25, random_state=103)
```

```python
# develop a function for evaluating the model

def print_evaluate(true, predicted):
    f1=np.round(f1_score(true, predicted, average='macro'),3)
    print('f1:', f1)
    
def cross_val(model, X, y):
    pred=cross_val_score(model, X,y,cv=kfold)
    return np.round(pred.mean(),3)

def evaluate(true, predicted):
    f1=np.round(f1_score(true, predicted, average='macro'),3)
    
    return f1
```

```python
log_reg=LogisticRegression(max_iter=1000)
log_reg.fit(X_train,y_train)
```

```python
train_pred = log_reg.predict(X_train)

print('Train set evaluation:\n')
print_evaluate(y_train, train_pred)
print('--------------------------------\n')
cross_score=cross_val(log_reg, X_train, y_train)
print('Cross_validation score:\n')
cross_score
```

```python
results_df=pd.DataFrame(data=[['Logistic Regression, l2', evaluate(y_train, train_pred), cross_score]], columns=['Model', 'f1', 'Cross Validation'])
results_df
```

```python
# we won't try another regularizers since not all regularizers are compatible with corresponding solver 
# and we are not supposed to use 'saga' and 'sag' (used with l1 and elastic net) solvers since we don't have scaled data
```
